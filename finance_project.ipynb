```
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/Peeyush2/financeBuddy/master/finance_data.csv'
df1 = pd.read_csv(url)
```

```
df1
```

```
#date formatting
# no inconsistencies in date
import datetime

dates = df1['transaction_date'].tolist()

allErrorDates = []

for completedDate in dates:
  completeDate = ''
  try:
    completeDate = datetime.datetime.strptime(completedDate, "%d-%m-%Y")
  except:
    completeDate = datetime.datetime.strptime(completeDate).strftime("%d-%m-%Y")
    allErrorDates.append(completeDate)

print(allErrorDates)
```
